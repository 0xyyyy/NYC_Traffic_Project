```python
#Gathers latitude and longitude data from googles geocoding api and stores it into the traffic_df so that we can display map in tableau for future use 
for i, row in traffic_df.iterrows():
    street1 = row["Roadway_Name"]
    street2 = row["From"]
    latlng_data = requests.get(f"{base_url}address={street1},{street2}+NY&key={api_key}").json()
    try:
        print(f"Gathering lat and long for {street1}, {street2}")
        latitude = latlng_data["results"][0]["geometry"]["location"]["lat"]
        longitude = latlng_data["results"][0]["geometry"]["location"]["lng"]
        traffic_df.loc[i, 'latitude'] = latitude
        traffic_df.loc[i, 'longitude'] = longitude
    except (KeyError, IndexError):
        print(f"could not find lat and long for {street1}, {street2}")
```

```python
#Need to create connection to sqlite file. Also will need to input latitude and longitude data into leaflet.js template?? 
conn = sqlite3.connect(db_file)

cursor = conn.cursor()

sql = '''CREATE TABLE trraffic(
    ID INT PRIMARY KEY,
    Entry_ID INT, 
    Segment_ID INT, 
    Roadway_Name CHAR, 
    From_ CHAR, 
    To_ CHAR,
    Direction CHAR,
    Date CHAR,
    time12_1AM INT,
    time1_2AM FLOAT,
    time2_3AM FLOAT, 
    time3_4AM FLOAT, 
    time4_5AM FLOAT, 
    time5_6AM FLOAT, 
    time6_7AM FLOAT, 
    time7_8AM FLOAT, 
    time8_9AM FLOAT, 
    time9_10AM FLOAT, 
    time10_11AM FLOAT, 
    time11_12PM FLOAT, 
    time12_1PM FLOAT, 
    time1_2PM FLOAT, 
    time2_3PM FLOAT, 
    time3_4PM FLOAT, 
    time4_5PM FLOAT, 
    time5_6PM FLOAT, 
    time6_7PM FLOAT, 
    time7_8PM FLOAT, 
    time8_9PM FLOAT, 
    time9_10PM FLOAT, 
    time10_11PM FLOAT,
    time11_12AM FLOAT,
    Latitude FLOAT, 
    Longitude FLOAT,
    Month INT,
    Day INT,
    Year INT
)'''

cursor.execute(sql)
print("Table created successfully__________________________________")

conn.commit()

conn.close()
```
